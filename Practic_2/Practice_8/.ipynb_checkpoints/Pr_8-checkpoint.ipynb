```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import time
import warnings

from sklearn.ensemble import BaggingRegressor
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
```

```python
data = pd.read_csv("Market_Basket_Optimisation.csv", header=None)
data
```

```python
item_count = data.apply(pd.Series.value_counts).sum(axis=1).sort_values(ascending=False)
item_count
```

```python
top_20 = item_count.head(20)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_20.index, y=top_20.values)
plt.xticks(rotation=90)
plt.xlabel("Items")
plt.ylabel("Frequency")
plt.title("Top 20 Most Popular Items")
plt.show()
```

```python
data.stack().value_counts(normalize=True)
```

```python
data.stack().value_counts(normalize=True).head(20).plot(kind='bar')
```

```python
data.stack().value_counts().apply(lambda item: item / data.shape[0]).head(20).plot(kind='bar')
```

```python
transactions = []
for i in range(data.shape[0]):
    row = data.iloc[i].dropna().tolist()
    transactions.append(row)
```

```python
from apriori_python import apriori
```

```python
t=[]
start=time.perf_counter()
t1, rules = apriori(transactions,minSup = 0.02, minConf = 0.3)
time1=(time.perf_counter()-start)
t.append(time1)
```

```python
rules
```

```python
from apyori import apriori
```

```python
start=time.perf_counter()
rules = apriori(transactions=transactions, min_support=0.02, min_confidence=0.3, min_lift=0.9)
results = list(rules)
time2=(time.perf_counter()-start)
t.append(time2)
```

```python
results
```

```python
for result in results:
    for subset in result[2]:
        print(subset[0], subset[1])
        print("Support: {0}; Confidence: {1}; Lift: {2};".format(result[1],subset[2],subset[3]))
        print()
```

```python
from efficient_apriori import apriori
```

```python
start=time.perf_counter()
itemsets, rules = apriori(transactions, min_support = 0.02, min_confidence = 0.3)
time3=(time.perf_counter()-start)
t.append(time3)
```

```python
for i in range(len(rules)):
    print(rules[i])
```

```python
print(time1, time2, time3)
```

```python
from fpgrowth_py import fpgrowth
```

```python
start=time.perf_counter()
itemsets, rules = fpgrowth(transactions, minSupRatio = 0.02, minConf= 0.3)
time4=(time.perf_counter()-start)
t.append(time4)
for i in range(len(rules)):
    print(rules[i])
```
